```python
import pandas as pd
import numpy as np

import statsmodels. api as sm
import statsmodels. formula. api as smf
```

```python
df = pd.read_csv('cars.csv')
print(f'{df.shape}')
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df['company'] = df.CarName.str.split().str.get(0)
df.head()
```

```python
df.company.unique()
```

```python
df['company'] = df.company.str.replace('maxda', 'mazda')
df['company'] = df.company.str.replace('Nissan', 'nissan')
df['company'] = df.company.str.replace('porcshce', 'porsche')
df['company'] = df.company.str.replace('toyouta', 'toyota')
df['company'] = df.company.str.replace('vokswagen', 'volkswagen')
df['company'] = df.company.str.replace('vw', 'volkswagen')
```

```python
df.company.unique()
```

```python
df.company.nunique()
```

```python
cars = df[['company', 'fueltype', 'aspiration', 'carbody', 'drivewheel',
         'wheelbase', 'carlength', 'carwidth', 'curbweight', 'enginetype',
         'cylindernumber', 'enginesize', 'boreratio', 'horsepower', 'price']]

cars.head()
```

```python
cars.corr()['price'].sort_values(ascending=False).round(2).to_frame()
```

```python
categorical = [i for i in cars.columns if cars[i].dtype not in ['int64','float64']]
categorical
```

```python
df_dummy = pd.get_dummies(data=cars[categorical], drop_first = True)
df_dummy.head()
```

```python
numeric = [i for i in cars.columns if cars[i].dtype in ['int64','float64']]
full_cars = pd.concat([cars[numeric], df_dummy], axis=1)
full_cars.head()
```

```python
results = smf.ols('price ~ horsepower', full_cars).fit()
print(results.summary())
```

```python
X1 = full_cars.drop(columns='price')
Y = full_cars['price']

X_1 = sm.add_constant(X1)
model2 = sm.OLS(Y, X_1)
results2 = model2.fit()
print(results2.summary())
```

```python
X2 = full_cars[[i for i in X1.columns if not i.startswith('company')]]

X_2 = sm.add_constant(X2)
model3 = sm.OLS(Y, X_2)
results3 = model3.fit()
print(results3.summary())
```

**Выбранная модель объясняет примерно 90% дисперсии. Среди предикторов 10 из 27 оказались не значимыми ($p > 0.05$). Пример интерпретации: при единичном изменении показателя horsepower, цена ВОЗРАСТАЕТ на 86.8164.**
