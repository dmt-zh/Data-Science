### Вопрос 1. Есть 10 различных марок шоколада. Какое количество способов подарить трём девочкам по шоколадке?


### Решение:


Каждая девочка может получить шоколадку любой марки. Поскольку все девочки и все марки различны, то количество способов подарить шоколадки девочкам равно количеству размещений с повторениями $ \overline{A}_{10}^3 = 10^3  $

```python
print('Количество способов подарить трём девочкам по шоколадке - {0:,}.'.format(10 ** 3).replace(',', ' '))
```

### Вопрос 2. Сколькими способами можно составить расписание авиарейсов на завтра, если всего имеется рейсы в 7 городов, а в день осуществляется от двух до четырех перелётов в разные города?


### Решение:

Если в день осуществляется только два рейса, то первый можно выбрать 7 способами, а второй уже только 6 способами. То есть расписание из двух рейсов можно составить $ A_7^2 $ способами. Аналогично, если в день состоится 3 рейса, то расписание можно составить $ A_7^3 $ способами, а для 4 рейсов в день − $ A_7^4 $.  По правилу сложения получаем $ A_7^2 + A_7^3 + A_7^4 $.

```python
a = (7 * 6) + int(factorial(7)/factorial(3)) + int(factorial(7)/factorial(4))
print('Количество способов составить расписание - {0:,}.'.format(a).replace(',', ' '))
```

### Вопрос 3. Сколькими способами можно вручить призы в 5 различных номинациях (в каждой номинации только один приз), если в соревновании участвуют 10 человек?


### Решение:


Приз в любой из 5 номинаций может достаться любому из 10 участников соревнований, то есть количество способов выдать призы равно количеству размещений с повторениями $ \overline{A}_{10}^5 $

```python
print('Количество способов вручить призы в 5 различных номинациях - {0:,}.'.format(10 ** 5).replace(',', ' '))
```

### Вопрос 4. Сколькими способами можно вручить призы в 5 различных номинациях (в каждой номинации только один приз), если в соревновании участвуют 10 человек, при условии, что каждый участник может получить не более одного приза?


### Решение:

Приз в первой номинации может достаться любому из 10 участников соревнований, во второй − любому из 9 участников (получивший первый приз другие призы получать не может), в третьей − любому из 8 участников, и т.д. То есть количество способов выдать призы в точности равно количеству размещений без повтореий $ A_{10}^5 $

```python
a = int(factorial(10)/factorial(5))
print('Количество способов вручить призы в 5 различных номинациях - {0:,}.'.format(a).replace(',', ' '))
```

### Вопрос 5. В семье семеро детей: старший − мальчик, дальше − девочка, девочка, мальчик, мальчик, мальчик и младшая − девочка. Сколькими способами родители могут выбрать имена, если они выбирают из 10 мужских и 13 женских имен и хотят, чтобы имена не повторялись?


### Решение:

Количество способов выбрать имена сыновьям равно количеству размещений без повторений $ A_{10}^4 $, аналогично, количество способов выбрать имена дочерям равно $ A_{13}^3 $. По правилу умножения получаем, что общее количество способов выбрать имена детям равно $ A_{10}^4 \cdot A_{13}^3 $

```python
a = int(factorial(10)/factorial(6) * factorial(13)/factorial(10))
print('Количество способов выбрать имена детям - {0:,}.'.format(a).replace(',', ' '))
```

### Вопрос 6. Жених и невеста выбирают трехъярусный свадебный торт. На выбор имеются 5 типов ярусов (бисквитный, йогуртовый, чизкейк и т.д.). Сколько различных тортов может предложить кондитер, если бисквитных ярусов может быть не больше двух, а ярусов  любого другого типа не больше одного?


### Решение:


Нужно рассмотреть 3 случая: в торте нет бисквитных ярусов, ровно один бисквитный ярус и ровно два бисквитных яруса.

- В первом случае (бисквитный ярус использовать нельзя) количество способов испечь торт равно количеству размещений без повторений $ A_4^3 $
- Во втором случае (ровно один бисквитный ярус) нужно сначала разместить бисквитный ярус, для это всего 3 варианта, а затем выбрать оставшиеся два яруса из 4 типов и определить для них места, то есть $ A_4^2 $. Всего получаем $ A_4^2 \cdot 3 $ тортов.
- В третьем случае (ровно два бисквитных яруса) нужно выбрать место для "небисвкитного" яруса на любом из 3 уровней, а затем выбрать тип для этого яруса из оставшихся 4 типов $ A_4^1 $. Всего получаем $ A_4^1 \cdot 3 $ тортов.

По правилу сложения получаем $ A_4^3 + (A_4^2 \cdot 3) + (A_4^1 \cdot 3) $

```python
a = int(factorial(4)/factorial(1)) + 3 * int(factorial(4)/factorial(2)) + 3 * 4
print(f'Количество различных тортов - {a}.')
```

### Вопрос 9. Группа из 8 студентов пришла в столовую. Сколькими способами они могут занять очередь друг за другом, если Маша и Таня хотят стоять рядом, а Коля не хочет быть последним?


### Решение:


Так как Маша и Таня хотят стоять рядом, то можно считать, что они занимают одно место вдвоём. Соответственно, мест становится 7, но надо учесть, что девочки могут между собой поменяться местами (Маша-Таня и Таня-Маша). Поэтому количество всех очередей надо будет удвоить.

Коля не хочет быть последним, поэтому для него есть 6 возможных мест в очереди. Остальные могут занять места 6! способами.

В итоге получаем:

```python
print('Количество способов занять очередь равно - {0:,}.'.format(2 * 6 * factorial(6)).replace(',', ' '))
```
