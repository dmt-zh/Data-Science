```python
import scipy as sc
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return math.sin(x/5.)*math.exp(x/10.) + 5*math.exp(-x/2.)
print(f(2))
print(f(30))
print('......................................................................................')

from scipy import optimize
a = optimize.minimize(f, 2, method = 'BFGS')
print(a)
print('......................................................................................')

b = optimize.minimize(f, 30, method = 'BFGS')
print(b)

k = np.arange(1, 31, 0.01)
x = []
for v in k:
    x.append(v)

y = []
for i in x:
    i = f(i)
    y.append(i)
plt.plot(x, y)
```

```python
k = np.arange(1, 31, 0.01)
x = []
for v in k:
    x.append(v)

y = []
for i in x:
    i = f(i)
    y.append(i)
y_int = []
for j in x:
    j = f(j)
    y_int.append(int(j))
plt.plot(x, y)
plt.plot(x, y_int)
```

```python
def h(x):
    return int(f(x))
p = [(1, 30)]
c = optimize.differential_evolution(h, p)
print(c)
print('......................................................................................')

q = optimize.minimize(h, 30, method = 'BFGS')
print(q)
```
