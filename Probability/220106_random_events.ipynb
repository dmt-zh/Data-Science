### Вопрос 1. Абонент забыл пин-код к своей сим-карте, однако помнит, что он содержит три «пятёрки», а одна из цифр – то ли «семёрка», то ли «восьмёрка». Какова вероятность включить телефон с первой попытки?


### Решение:

$ C_{4}^1 $ способами можно выбрать место, на котором расположена сомнительная цифра и на каждом из этих четырёх мест могут располагаться 2 цифры (семёрка или восьмёрка). По правилу сложения комбинаций, общее число исходов $ C_{4}^1 + C_{4}^1 = 4 + 4 = 8 $. Благоприятствующий исход один (правильный пин-код). 

Таким образом, по классическому определению:

```python
print(f'Вероятность включить телефон с 1-й попытки  - {1/8}')
```

### Вопрос 2. У Васи дома живут четыре кота, каждый из которых после завтрака случайным образом и с равной вероятностью может выйти на прогулку или остаться дома. Какова вероятность того, что завтра после завтрака уйдут гулять ровно 2 кота.


### Решение:


Каждый из четырех котов может пойти на прогулку или остаться дома. То есть общее число всех возможных исходов равно количеству размещений с повторениями $ \overline{A}_{4}^2 = 4^2 $.
Количество способов составить комбинацию из двух котов, которые уйдут на прогулку $ C_{4}^2 = \frac{4!}{2! \cdot 2!} $.

Таким образом, по классическому определению:

```python
m = comb(4, 2)
n = 4 ** 2
print(f'Вероятность того, что завтра после завтрака уйдут гулять ровно 2 кота  - {m/n}')
```

### Вопрос 3. В ящике находится 15 качественных и 5 бракованных деталей. Наудачу извлекаются 2 детали. Найти вероятность того, что:

- **а) обе детали будут качественными;**
- **б) одна деталь будет качественной, а одна – бракованной;**
- **в) обе детали бракованны.**


### Решение:

Вычислим общее число исходов:  $ C_{20}^2 = \frac{20!}{2! \cdot 18!} $ способами можно извлечь 2 детали из ящика.

- а) Количество способов извлечь 2 качественные детали:  $ C_{15}^2 = \frac{15!}{2! \cdot 13!} $
- б) Количество способов извлечь 1 качественныу деталь и 1 бракованную:  $ C_{15}^1 \cdot C_{5}^1 $
- в) Количество способов извлечь 2 бракованные детали:  $ C_{5}^2 = \frac{5!}{2! \cdot 3!} $

Таким образом, по классическому определению:

```python
n = comb(20, 2)
ma = comb(15, 2)
mb = comb(15, 1) * comb(5, 1)
mc = comb(5, 2)

print(f'Вероятность того, что обе детали будут качественными  - {ma/n:.4}')
print(f'Вероятность того, что одна деталь будет качественной, а одна бракованной  - {mb/n:.4}')
print(f'Вероятность того, что обе детали бракованны  - {mc/n:.4}')
```

### Вопрос 4. Студент знает ответы на 25 экзаменационных вопросов из 60. Какова вероятность сдать экзамен, если для этого необходимо ответить не менее чем на 2 из 3 вопросов?



### Решение:

Вычислим общее число исходов:  $ C_{60}^3 = \frac{60!}{3! \cdot 57!} $ способами можно выбрать 3 вопроса из 60.

Для того чтобы сдать экзамен, нужно ответить на 2 или 3 вопроса:
- Количество способов выбрать 2 «хороших» вопроса и один «плохой» $ C_{25}^2 \cdot C_{35}^1 $
- Количество способов выбрать 3 «хороших» вопроса $ C_{25}^3 $

По правилу сложения комбинаций, общее число благоприятствующее для сдачи экзамена исходов: $ C_{25}^2 \cdot C_{35}^1 + C_{25}^3 $

Таким образом, по классическому определению:

```python
n = comb(60, 3)
m = comb(25, 2) * comb(35, 1) + comb(25, 3)
print(f'Вероятность того, что что студент сдаст экзамен  - {m/n:.4}')
```
