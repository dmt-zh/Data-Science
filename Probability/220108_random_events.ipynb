### Вопрос 1. После бури на участке между 40-м и 70-м километрами телефонной линии произошёл обрыв провода. Какова вероятность того, что он произошёл между 50-м и 55-м километрами линии?


### Решение:

Общему числу исходов соответствует участок длиной, $ L = 70 - 40 = 30 км $, благоприятствующему количеству исходов – участок длиной $ l = 55 - 50 = 5 км $.

Таким образом, по геометрическому определению вероятности:

```python
print(f'Вероятность того, что обрыв провода произошёл между 50-м и 55-м километрами линии  - {5/30:.4}')
```

### Вопрос 2. В круге радиуса 10 см находится прямоугольный треугольник с катетами 12 и 7 см. В круг наудачу ставится точка. Найти вероятность того, что она не попадёт в данный треугольник.


### Решение:

Общему количеству исходов соответствует площадь круга: $ S_{k} = \pi r^2 $ <br>
Площадь прямоугольного треугольника равна полупроизведению его катетов: $ S_{t} = \frac{1}{2} \cdot ab $ <br>
По условию поставленная в круг точка **не должна** попасть в треугольник, поэтому благоприятствующее число исходов выражается разностью $ S_{k} - S_{t} $

Таким образом, по геометрическому определению вероятности:

```python
Sk = pi * 100
St = (12 * 7)/2
print(f'Вероятность того, что поставленная в круг точка не попадёт в треугольник  - {(Sk - St)/Sk:.4}')
```

### Вопрос 3. Студенты случайным образом приходят в столовую с 14.00 до 15.00, при этом обед каждого из них занимает примерно 20 минут. Найти вероятность того, что: 
- **а) Коля встретится с Олей во время обеда**
- **б) данная встреча не состоится**


### Решение:

Оля и Коля могут встретиться в течение 60 минут. Построим квадрат размером 60 на 60 единиц.

"https://github.com/dmt-zh/Data-Science/blob/main/Probability/square.PNG"

Площадь квадрата соответствует общему числу исходов: $ S = 60^2 $ <br>
Рассмотрим противоположные события:
- $ A $ Оля и Коля встретятся во время обеда;
- $ \overline{A} $ данной встречи не состоится.

Множеству благоприятствующих исходов (когда Оля и Коля встретятся) соответствует площадь заштрихованной фигуры. Вычислим площади двух прямоугольных треугольников (по формуле $ S_{t} = \frac{1}{2} \cdot ab $) и из площади квадрата вычитем площади треугольников, получая тем самым благоприятствующую площадь: $ \tilde{S} = S - (S_{t1} + S_{t2}) $ <br>
Противоположные события образуют полную группу, по соответствующей теореме: $ p = 1 - q $

Таким образом, по геометрическому определению вероятности:

```python
S = 60 ** 2
S_t1 = (40 * 40)/2
S_t2 = (40 * 40)/2
Sa = S - (S_t1 + S_t2)

print(f'Вероятность того, Оля и Коля встретятся во время обеда  - {Sa/S:.4}')
print(f'Вероятность того, данной встречи не состоится  - {1- (Sa/S):.4}')
```
