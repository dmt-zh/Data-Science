### Вопрос 1. В трех урнах имеется по 6 белых и по 4 черных шара. Из каждой урны извлекают наудачу по одному шару. Найти вероятность того, что: 
- **а) все три шара будут белыми**
- **б) все три шара будут одного цвета**


### Решение:

Рассмотрим события: – $A_1 , A_2 , A_3 $ из 1-й, 2-й и 3-й урны соответственно будет извлечён белый шар. По классическому определению вероятности: $ P( A1 ) = P( A2 ) = P( A3 ) = \frac{6}{10} $. Тогда вероятности извлечения чёрного шара из соответствующих урн равны: $ P(\overline{A}_{1}) = P(\overline{A}_{2}) = P(\overline{A}_{3})= 1 - \frac{6}{10} $.

- а) Рассмотрим событие: $ B $ – из каждой урны будет извлечено по одному белому шару. Данное событие выражается в виде произведения $ B = A_1 \cdot A_2 \cdot A_3 $. По теореме умножения вероятностей независимых событий: $ P(B) = P( A_1 A_2 A_3 ) = P( A_1 ) \cdot P( A_2 ) \cdot P( A_3 ) $
- б) Рассмотрим событие $ C $ – из каждой урны будет извлечено по 1 одному чёрному шару. По теореме умножения вероятностей независимых событий: $ P(C) = P( \overline{A}_{1} \overline{A}_{2} \overline{A}_{3} ) = P( \overline{A}_{1} ) \cdot P( \overline{A}_{2} ) \cdot P( \overline{A}_{3} ) $. Рассмотрим событие D – все три шара будут одного цвета. Данное событие $ D = B + C $ состоит в двух несовместных исходах: (будут извлечены 3 белых или 3 чёрных шара).

Таким образом:

```python
Pa = 6/10
Pb = Pa ** 3
Pd = Pb + (1 - Pa) ** 3

print(f'Вероятность того, что все три шара будут белыми  - {Pb:.3}')
print(f'Вероятность того, что все три шара будут одного цвета  - {Pd:.3}')
```

### Вопрос 2. Для сигнализации о возгорании установлены два независимо работающих датчика. Вероятности того, что при возгорании датчик сработает, для первого и второго датчиков соответственно равны 0,5 и 0,7. Найти вероятность того, что при пожаре:
- **а) оба датчика откажут**
- **б) оба датчика сработают**
- **в) сработает только один датчик**


### Решение:


Рассмотрим события: – $A_1$ – при возгорании сработает 1-й датчик, $A_2$ – при возгорании сработает 2-й датчик. По условию $P(A1)$ = 0,5 и $P(A2)$ = 0,7. Вероятности противоположных событий: $ P(\overline{A}_{1}) = 1 - 0,5 = 0,5 $ и $ P(\overline{A}_{2}) = 1 - 0,7 = 0,3$.

- а) Рассмотрим событие: $ B $ – при пожаре оба датчика откажут. По теореме умножения вероятностей независимых событий: $ B = P( \overline{A}_{1} ) \cdot P( \overline{A}_{2} ) $
- б) Рассмотрим событие $ C $ – при пожаре оба датчика сработают. По теореме умножения вероятностей независимых событий:  $ C = P( A1 ) \cdot P( A2 ) $
- в) Рассмотрим событие: $D$ – при пожаре сработает только один датчик. События образуют полную группу, следовательно: $ P(B) + P(C) + P(D) = 1  \Rightarrow   P(D) = 1 - P(B) - P(C) $


Таким образом:

```python
Pb = 0.5 * 0.3
Pc = 0.5 * 0.7
Pd = 1 - Pb - Pc

print(f'Вероятность того, что оба датчика откажут  - {Pb}')
print(f'Вероятность того, что оба датчика сработают  - {Pc}')
print(f'Вероятность того, что сработает только один датчик  - {Pd}')
```

### Вопрос 3. Вероятность поражения цели первым стрелком при одном выстреле равна 0,8. Вероятность того, что цель не поражена после выполнения первым и вторым стрелками по одному выстрелу равна 0,08. Какова вероятность поражения цели вторым стрелком при одном выстреле?


### Решение:

По условию $ p_1 = 0,8 $, – вероятность поражения цели первым стрелком при одном выстреле. Тогда вероятность его промаха: $ q_1 = 1 - p_1 $. Обозначим через $ p_2 , q_2 $– вероятности попадания и промаха 2-го стрелка. По теореме умножения вероятностей независимых событий: $ q = q_1 \cdot q_2 $ – вероятность того, что оба стрелка промахнутся $ \Rightarrow  q_2 = \frac{q}{q_1} $. В результате: $ p_2 = 1 - \frac{q}{q_1} $.

Таким образом:

```python
print(f'Вероятность поражения цели вторым стрелком при одном выстреле  - {1 - (0.08/0.2):.2}')
```

### Вопрос 3. Из трех орудий произвели залп по цели. Вероятность попадания при одном выстреле только из первого орудия равна 0,7, из второго – 0,6, из третьего – 0,8. Найти вероятность того, что: 
- **1) хотя бы один снаряд попадет в цель**
- **2) только два снаряда попадут в цель** 
- **3) цель будет поражена не менее двух раз**

### Решение:

По условию $ p_1 = 0,7 $, $p_2 = 0,6$ и $ p_3 = 0,8 $, – вероятности попадания в цель из соответствующих орудий. Тогда соответствующие вероятности промаха: $ q_1 = 1 - p_1, q_2 = 1 - p_2, q_3 = 1 - p_3 $.
- 1) По теореме умножения вероятностей независимых событий: $ p_{(0)} = q_1q_2q_3 $ – вероятность того, что будет три промаха. Тогда:  $ p_{(1 , 2 , 3)} = 1 - p_{(0)} $ – вероятность того, что хотя бы один снаряд попадет в цель. 
- 2) Событие «только два снаряда попадут в цель» состоит в трёх несовместных исходах: попадание из 1-го и 2-го орудий и промах из 3-го или попадание из 1-го и промах из 2-го и попадание из 3-го орудия или промах из 1-го и попадание из 2-го и 3-го орудий. По теоремам сложения вероятностей несовместных и умножения вероятностей независимых событий: $ p_{(2)} = p_1p_2q_3 + p_1q_2p_3 + q_1 p_2 p_3 $
- 3) ) По теореме умножения вероятностей независимых событий: $ p_{(3)} = p_1 p_2 p_3 $ – вероятность того, что все три снаряда попадут в цель. По теореме сложения вероятностей несовместных событий: $ p_{(2 , 3)} = p_{(2)} + p_{(3)} $ – вероятность того, что цель будет поражена не менее двух раз.

Таким образом:

```python
p_123 = 1 - (0.3 * 0.4 * 0.2)
p_2 = (0.7 * 0.6 * 0.2) + (0.7 * 0.4 * 0.8) + (0.3 * 0.6 * 0.8)
p_3 = 0.7 * 0.6 * 0.8

print(f'Вероятность того, что хотя бы один снаряд попадет в цель - {p_123}')
print(f'Вероятность того, что только два снаряда попадут в цель - {p_2:.4}')
print(f'Вероятность того, что цель будет поражена не менее двух раз - {p_2 + p_3}')
```
