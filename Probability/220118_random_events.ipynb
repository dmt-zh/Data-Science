### Вопрос 1. Производится 8 выстрелов по цели, в каждом из которых вероятность попадания равна 0,1. Для разрушения цели требуется хотя бы два попадания. Найти вероятность того, что цель будет разрушена.


### Решение:

Используем формулу Бернулли: $ P_{n}^m = C_{n}^m \cdot p^m \cdot q^{n-m} $, в данном случае:
- $ n = 8$ - всего выстрелов;
- $ p = 0.1$ - вероятность попадания в цель при каждом выстреле;
- $ q = 1 - p = 1 - 0.1 = 0.9$ - вероятность промаха при каждом выстреле.

По теореме сложения вероятностей несовместных событий:<br>
$ P_8(m < 2) = P_8^0 + P_8^1 = C_8^0 \cdot 0.1^0 \cdot 0.9^8 + C_8^1 \cdot 0.1^1 \cdot 0.9^7 $ - вероятность того, что цель не будет разрушена (ни одного или 1 попадание);

Найдём вероятность противоположного события: $ P_8(m \geq 2) = 1 - P_8(m < 2) $. Таким образом:

```python
P8 = 0.9 ** 8 + 8 * 0.1 * 0.9 ** 7
print(f'Вероятность того, что цель будет разрушена  - {1 - P8:.4}')
```

### Вопрос 2. Монета подбрасывает 9 раз. Найти вероятность наивероятнейшего числа появлений орла.


### Решение:

В данной задаче речь идёт о независимых испытаниях, при этом:
- $ n = 9 $ - всего испытаний;
- $ p = 0,5 $ - вероятность выпадения орла в каждом испытании;
- $ q = 1 - p = 1 - 0,5 = 0,5 $ - вероятность выпадения решки в каждом испытании.

Найдём наивероятнейшее количество $ m_0 $ появлений орла:<br>
$ np - q \leq  m_0 < np + p $<br>
$ 9 \cdot 0.5 - 0.5 \leq  m_0 <9 \cdot 0.5 + 0.5 $<br>
$ 4 \leq  m_0 < 5 $<br>

Так как $ np - q = 4 $ - целое число, то существуют два наивероятнейших значения: $ m_0 = 4 $ и $ m_0 = 5 $. Используя формулу Бернулли, вычислим соответствующие вероятности: $ P_9^4 = P_9^5 = C_9^5 \cdot 0.5^5 \cdot 0.5^4 $. Таким образом:

```python
P = comb(9, 5) * 0.5 ** 9
print(f'Наивероятнейшее количество появлений орла  4 и 5 раз.')
print(f'Вероятность наивероятнейшего числа появлений орла  - {P:.4}')
```

### Вопрос 3. Вероятность изготовления бракованных деталей при их массовом производстве равна 0,001. Определить вероятность того, что в партии из 800 деталей будет:
- **а) ровно 2 бракованные;**
- **б) не более двух.**


### Решение:


Используем формулу Пуассона: $ P_{m} \approx \frac{\lambda^m}{m!} \cdot e^{-\lambda} $, в данном случае: $\lambda = np = 800 \cdot 0.001 = 0.8 $
- а) $ P_{2} \approx \frac{0.8^2}{2!} \cdot e^{-0.8}$
- б) По теореме сложения вероятностей несовместных событий:<br>
$ P(m \geq 2) \approx P_{0} + P_{1} + P_{2} \approx \frac{0.8^0}{0!} \cdot e^{-0.8} + \frac{0.8^1}{1!} \cdot e^{-0.8} + \frac{0.8^2}{2!} \cdot e^{-0.8} $<br>

Таким образом:

```python
L = 800 * 0.001
a = e**-L

P2 = (L ** 2 / 2) * a
P1 = L * a
P0 = a
P012 = sum([P0, P1, P2])

print(f'Вероятность того, что в данной партии окажется ровно две бракованные детали  - {P2:.4}')
print(f'Вероятность того, что в данной партии окажется не более двух бракованных изделий - {P012:.4}')
```
